5. GEOSPATIAL ANALYSIS 

includes latitude and longitude, you can plot the accident locations on a map.


```python
import folium
from folium.plugins import HeatMap

# Sample 1% of the dataset
sample_df = df.sample(int(0.01 * len(df)))

# Create latitude and longitude pairs
lat_lon_pairs = list(zip(list(sample_df.Start_Lat), list(sample_df.Start_Lng)))

# Initialize the map at the center of the dataset (average lat/lon)
map = folium.Map(location=[df.Start_Lat.mean(), df.Start_Lng.mean()], zoom_start=10)

# Add heatmap to the map
HeatMap(lat_lon_pairs).add_to(map)

# Display the map
map
```
